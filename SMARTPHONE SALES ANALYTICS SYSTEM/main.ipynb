```python
def printphone():

    data = pd.read_csv(r"C:\Users\shobh\Downloads\mobile_dataset_100_rows.csv")
    
    df = pd.DataFrame(data)
    
    console = Console()
    
    table = Table(title=" Mobile Dataset")
    
    for column in df.columns:
        table.add_column(column, style="cyan")
    
    for index, row in df.iterrows():
        table.add_row(
            str(row["Brand"]),
            str(row["Model"]),
            str(row["Price"]),
            str(row["Rating"]),
            str(row["Battery"]),
            str(row["RAM"]),
            str(row["Storage"]),
            str(row["Sales"]),
            str(row["Launch Year"])
        )
    
    console.print(table)


def choose2():

    result = df[df["Brand"] == search]
    
    console = Console()
    
    table = Table(title="Search Result")
    
    for column in result.columns:
        table.add_column(column, style="green")
    
    for index, row in result.iterrows():
        table.add_row(
            str(row["Brand"]),
        str(row["Model"]),
        str(row["Price"]),
        str(row["Rating"]),
        str(row["Battery"]),
        str(row["RAM"]),
        str(row["Storage"]),
        str(row["Sales"]),
        str(row["Launch Year"])
    )

    console.print(table)


def choose3():

    fltr = df[
        (df["Price"] > minimum) &
        (df["Price"] < maximum)
    ]
    
    final = fltr.sort_values(by="Price", ascending=True)
    
    console = Console()
    
    table = Table(title=" Mobile Phones Within Budget Range")
    
    for column in final.columns:
        table.add_column(column, style="cyan", justify="center")
    
    for index, row in final.iterrows():
        table.add_row(
            str(row["Brand"]),
            str(row["Model"]),
        str(row["Price"]),
        str(row["Rating"]),
        str(row["Battery"]),
        str(row["RAM"]),
        str(row["Storage"]),
        str(row["Sales"]),
        str(row["Launch Year"])
    )
    
    console.print(table)




def choose4():

    fltr = df.sort_values(by="Rating", ascending=False)
    
    top_5 = fltr.head(5)
    
    console = Console()
    
    table = Table(title="Top 5 Highest Rated Mobile Phones")
    
    for column in top_5.columns:
        table.add_column(column, style="magenta", justify="center")
    
    for index, row in top_5.iterrows():
        table.add_row(
            str(row["Brand"]),
            str(row["Model"]),
            str(row["Price"]),
            str(row["Rating"]),
            str(row["Battery"]),
            str(row["RAM"]),
            str(row["Storage"]),
            str(row["Sales"]),
            str(row["Launch Year"])
        )
    
    console.print(table)

def choose5():

    fltr = df.sort_values(by="Sales", ascending=False)
    
    top_5 = fltr.head(5)
    
    console = Console()
    
    table = Table(title=" Top 5 Best Selling Mobile Phones")
    
    for column in top_5.columns:
        table.add_column(column, style="yellow", justify="center")
    
    for index, row in top_5.iterrows():
        table.add_row(
            str(row["Brand"]),
            str(row["Model"]),
            str(row["Price"]),
            str(row["Rating"]),
            str(row["Battery"]),
            str(row["RAM"]),
            str(row["Storage"]),
            str(row["Sales"]),
            str(row["Launch Year"])
        )

    console.print(table)

def choose6():

    console = Console()
    
    grp = df.groupby("Brand")["Battery"].mean().round(0)
    
    table1 = Table(title=" Average Battery Capacity By Brand")
    
    table1.add_column("Brand", style="cyan", justify="center")
    table1.add_column("Average Battery", style="green", justify="center")
    
    for brand, battery in grp.items():
        table1.add_row(
            str(brand),
            f"{int(battery)} mAh"
        )
        
    console.print(table1)
    
    fltr = df.sort_values(by="Battery", ascending=False)
    
    fltr.index = range(1, len(fltr) + 1)
    
    top_mobile = fltr.head(1)
    
    table2 = Table(title= "Mobile With Highest Battery Capacity")
    
    for column in top_mobile.columns:
        table2.add_column(column, style="magenta", justify="center")

    for index, row in top_mobile.iterrows():
        table2.add_row(
            str(row["Brand"]),
            str(row["Model"]),
            str(row["Price"]),
            str(row["Rating"]),
            str(row["Battery"]),
            str(row["RAM"]),
            str(row["Storage"]),
            str(row["Sales"]),
            str(row["Launch Year"])
        )
    
    console.print(table2)

def avg_battery_by_brand():
    
    grp = df.groupby("Brand")["Battery"].mean().round(0)

    table = Table(title=" Average Battery Capacity By Brand")

    table.add_column("Brand", style="cyan", justify="center")
    table.add_column("Average Battery", style="green", justify="center")

    for brand, battery in grp.items():
        table.add_row(
            str(brand),
            f"{int(battery)} mAh"
        )

    console.print(table)
```

```python

while True:
    print(menu)
    choose=int(input("Enter your choice:"))

    if choose==1:
        printphone()
    elif choose==2:
        search = input("Enter phone brand:").strip().title()
        choose2()
    elif choose==3:
        minimum = int(input("Enter Minimum Amount:"))
        maximum = int(input("Enter Maximum Amount:"))
        choose3()
    elif choose==4:
        choose4()
    elif choose==5:
        choose5()
    elif choose==6:
        choose6()
    elif choose==7:
        print(dash)
        opt = int(input("Select an option :"))
        if opt ==1:
            xy = df.groupby("Brand")["Sales"].mean()
            plt.figure(figsize = (10,10))
            plt.bar(xy.index , xy.values)
            plt.title("Brand-wise Sales Chart")
            plt.ylabel("Sales Per Unit")
            plt.xlabel("Brands")
            plt.show()
        elif opt==2:
            x = df["Price"]
            y = df["Rating"]
            plt.figure(figsize=(10,10))
            plt.scatter(x,y)
            plt.title("Price vs Rating Scatter Plot")
            plt.ylabel("Ratings")
            plt.xlabel("Price")
            plt.show()
        elif opt==3:
            rate = df["Rating"]
            plt.figure(figsize=(8,8))
            plt.title("Rating Distribution")
            plt.hist(rate)
            plt.show()
        elif opt==4:
            grp = df.groupby("Brand")["Battery"].mean().round(1)
            plt.figure(figsize=(10,7))
            plt.title("Battery Comparison Chart")
            plt.bar(grp.index,grp.values)
            plt.ylabel("Battery Capacity (mAh)")
            plt.xlabel("Brands")
            plt.show()
        elif opt==5:
            count = df["Storage"].value_counts()

            ind = []
            val = []
            
            for i in count.index:
                ind.append(str(i) + "GB")
            
            for x in count.values:
                val.append(x)
            
            plt.figure(figsize=(8,6),dpi=100)
            
            plt.bar(ind, val)
            
            plt.title("Storage Variant Analysis")
            plt.xlabel("Storage Variants")
            plt.ylabel("Number of Phones")
            
            plt.show()
        else:
            print("Invalid Operation Performed ! Please Enter From 1  To 5")



    else:
        print("Inalid Operation Performed ! Please Select From the menu ")
                    
```
