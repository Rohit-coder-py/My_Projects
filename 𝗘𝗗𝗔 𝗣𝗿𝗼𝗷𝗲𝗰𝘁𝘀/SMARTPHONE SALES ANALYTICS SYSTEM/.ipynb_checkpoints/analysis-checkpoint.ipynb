```python
data = pd.read_csv(r"C:\Users\shobh\Downloads\mobile_dataset_100_rows.csv")

df = pd.DataFrame(data)

pd.set_option("display.max_row",None)
df
```

```python
grp = df.groupby("Brand")["Battery"].mean()
grp


#highest battery 

fltr = df.sort_values(by = "Battery",ascending=False)
fltr.index = range(1,len(fltr)+1)
fltr.head(1)
```

```python
xy = df.groupby("Brand")["Sales"].mean()

plt.figure(figsize=(10,10))
plt.bar(xy.index,xy.values)

plt.ylabel("Unit Sales")
plt.xlabel("Smart Phones")
plt.title("Brand-wise Sales Chart")
```

```python
xy = df.groupby("Price")["Rating"].mean()
plt.figure(figsize=(10,10))
plt.scatter(xy.index,xy.values)

plt.ylabel("Unit Sales")
plt.xlabel("Smart Phones")
plt.title("Brand-wise Sales Chart")
```

```python
x = df["Price"]
y = df["Rating"]

plt.scatter(x,y)
```

```python
rate = df["Rating"]

plt.figure(figsize=(8,8))
plt.title("Rating Distribution")
plt.hist(rate)
plt.show()
```

```python
grp = df.groupby("Brand")["Battery"].mean().round(1)

plt.figure(figsize=(10,7))

plt.title("Battery Comparison Chart")
plt.bar(grp.index,grp.values)
plt.ylabel("MAH Battery")
plt.xlabel("Brands")
```

```python
df
```

```python
count = df["Storage"].value_counts()

ind = []
val = []

for i in count.index:
    ind.append(str(i) + "GB")

for x in count.values:
    val.append(x)

plt.figure(figsize=(8,6))

plt.bar(ind, val)

plt.title("Storage Variant Analysis")
plt.xlabel("Storage Variants")
plt.ylabel("Number of Phones")

plt.show()
```

```python
grp = df.groupby("Brand")["Sales"].sum()

plt.figure(figsize=(8,8))

plt.pie(grp.values, labels=grp.index, autopct="%1.1f%%")

plt.title("Smartphone Market Share")

plt.show()
```
