```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns 
```

<b><span style="background-color:#7C3AED; color:white; padding:6px 14px; border-radius:20px;"> DATA CLEANING TASKS</span></b>

<b><span style="background-color:#111827; color:#22C55E; padding:6px 14px; border:2px solid #22C55E; border-radius:10px;">Task 1</span></b>

```python
# Understand Dataset

# check rows & columns
# view sample data
# inspect column names
# understand dataset structure


data = pd.read_csv(r"C:\Users\shobh\OneDrive\Attachments\Documents\Desktop\py\covid_19_india.csv")

df = pd.DataFrame(data)

df.index = range(1,len(df)+1)

df

df.sample()

df.columns

```

```python
#Adding some columns mannually to get provide better insights 

df['Active'] = df['Confirmed'] - df['Cured'] - df['Deaths'].astype(int)
df['Recovery Rate'] = (df['Cured'] / df['Confirmed']) * 100
df['Death Rate'] = (df['Deaths'] / df['Confirmed']) * 100
df['Daily Cases'] = df['Confirmed'].diff()

```

<b><span style="background-color:#111827; color:#22C55E; padding:6px 14px; border:2px solid #22C55E; border-radius:10px;">Task 2</span></b>

```python
# 📌 Task 2
# Check Missing Values

# find null values
# identify problematic columns
# handle missing data


summ = df.isnull().sum()

print(summ)

```

```python
df[['Sno', 'Date', 'Time', 'State/UnionTerritory',
       'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured',
       'Deaths', 'Confirmed', 'Active', 'Recovery Rate', 'Death Rate',
       'Daily Cases']] = df[['Sno', 'Date', 'Time', 'State/UnionTerritory',
       'ConfirmedIndianNational', 'ConfirmedForeignNational', 'Cured',
       'Deaths', 'Confirmed', 'Active', 'Recovery Rate', 'Death Rate',
       'Daily Cases']].replace("-",0)
```

<b><span style="background-color:#111827; color:#22C55E; padding:6px 14px; border:2px solid #22C55E; border-radius:10px;">Task 3</span></b>

```python
# Remove Duplicate Rows

# detect duplicates
# remove duplicates if present


df[df.duplicated()]


#no rows = no delete 

#we can delete it using df.drop_duplicated()
```

<b><span style="background-color:#111827; color:#22C55E; padding:6px 14px; border:2px solid #22C55E; border-radius:10px;">Task 4</span></b>

```python
# Fix Data Types

# convert date column to datetime
# convert numeric columns properly

df.sample()
```

```python
# convert date column
df['Date'] = pd.to_datetime(df['Date'])

df[['ConfirmedIndianNational', 'ConfirmedForeignNational']] = (
    df[['ConfirmedIndianNational', 'ConfirmedForeignNational']]
    .astype(int)
)

# check datatypes
df.info()
```

```python
df[['Recovery Rate','Death Rate','Daily Cases']] = (df[['Recovery Rate','Death Rate','Daily Cases']].astype(float))
```

```python
df.info()
```

<b><span style="background-color:#111827; color:#22C55E; padding:6px 14px; border:2px solid #22C55E; border-radius:10px;">Task 5</span></b>

```python
# Clean Column Names

# rename columns
# remove spaces/special characters
```

```python
df.rename(columns = {"Sno": "Serial No."} ,inplace = True)

df.sample()
```

```python
# remove spaces/special characters


df['State/UnionTerritory'] = df['State/UnionTerritory'].str.strip()
```

<b><span style="background-color:#111827; color:#22C55E; padding:6px 14px; border:2px solid #22C55E; border-radius:10px;">Task 7</span></b>

```python
# Most Affected States

# Find:

# top states by confirmed cases
# Chart:

# horizontal barplot
grp = df.groupby('State/UnionTerritory')['Confirmed'] \
        .sum() \
        .sort_values(ascending=False) \
        .head(10)

plt.figure(figsize=(20,10))

sns.barplot(
    x=grp.values,
    y=grp.index,
    palette='rainbow'
)

plt.title("Top 10 States by Confirmed Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("States")

plt.show()
```

<b><span style="background-color:#111827; color:#22C55E; padding:6px 14px; border:2px solid #22C55E; border-radius:10px;">Task 8</span></b>

```python
# 📌 Task 8
# Least Affected States

# Find:

# states with lowest confirmed cases
# Chart:

# barplot

grp = df.groupby('State/UnionTerritory')['Confirmed'] \
        .sum() \
        .sort_values(ascending=True) \
        .head(10)

plt.figure(figsize=(20,10))

sns.barplot(
    x=grp.values,
    y=grp.index,
    palette='rainbow'
)

plt.title("Top 10 States by Confirmed Cases")
plt.xlabel("Confirmed Cases")
plt.ylabel("States")

plt.show()
```

```python

# Daily New Cases Analysis

# Create:

# daily increase column
# Chart:

# lineplot

df.sample()
```

```python
daily = df.groupby('Date')['Confirmed'].sum()

new = daily.diff()

plt.figure(figsize=(15,6))

plt.plot(new.index, new.values)

plt.xticks(rotation=45)

plt.ticklabel_format(style='plain', axis='y')

plt.xlabel("Date")
plt.ylabel("Daily Cases")

plt.title("India Daily COVID Cases")

plt.show()
```
