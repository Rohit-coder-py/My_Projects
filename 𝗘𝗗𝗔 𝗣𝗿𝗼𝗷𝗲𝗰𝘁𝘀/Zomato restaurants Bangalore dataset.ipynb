<b><span style="background-color:#111827; color:#22C55E; padding:6px 14px; border:2px solid #22C55E; border-radius:10px;">Task 15 </span></b>

```python
# Analyze reviews vs ratings relationship

# do highly reviewed restaurants have higher ratings?
# correlation analysis

# Analyze Reviews vs Ratings Relationship

data = df[['Dinner Ratings', 'Dinner Reviews']].dropna()

data['Dinner Ratings'] = data['Dinner Ratings'].astype(float)
data['Dinner Reviews'] = data['Dinner Reviews'].astype(int)

# Correlation
corr = data['Dinner Ratings'].corr(data['Dinner Reviews'])

print("Correlation Between Reviews and Ratings :", corr)

# Visualization
plt.figure(figsize=(10,5))

sns.regplot(
    x=data['Dinner Reviews'],
    y=data['Dinner Ratings']
)

plt.title("Dinner Reviews vs Dinner Ratings")
plt.xlabel("Number of Reviews")
plt.ylabel("Ratings")

plt.show()

```
