```python
# Outlier Analysis

# Find:

# extremely slow deliveries
# extremely fast deliveries
# percentage of outliers
# impact of outliers on statistics
# Chart:

# boxplot
```

```python
q1 = data.quantile(0.25)
q3 = data.quantile(0.75)

iqr = q3-q1

print(iqr)

lower_fence = q1 -1.5 *iqr
upper_fence = q3 +1.5 *iqr

```

```python
 # fast deliveries
fast = df[df['Delivery_Time_min'] < lower_fence]

# unusually slow deliveries
slow = df[df['Delivery_Time_min'] > upper_fence]
```

```python
# percentage of outliers
total_outliers = len(slow) + len(fast)

percentage = (total_outliers / len(df)) * 100

print(percentage)
```

```python
# impact of outliers on statistics

clean_df = df[ (df['Delivery_Time_min']  >= lower_fence)     &     (df['Delivery_Time_min']  <=  upper_fence)]

final = clean_df['Delivery_Time_min']
```
