```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
data = pd.read_csv(r"C:\Users\shobh\Downloads\IPL.csv")

df = pd.DataFrame(data)
df.index = range(1,len(df)+1)

df
```

```python
df.info()

df.sample()

df.columns

df.describe()

df.shape
```

```python
df.head()
```

```python
win = df["match_winner"].value_counts()

plt.figure(figsize=(10,10))

sns.barplot(
    x=win.index,
    y=win.values,
    palette = 'rainbow'
)

plt.title("Match Wins by Team")

plt.show()
```

```python
x = df["toss_decision"]

fig = figsize = (10,10)

sns.countplot(x = x , palette = 'rainbow')

plt.title("Toss Decisions ")

plt.xlabel("Teams")
plt.ylabel("No. Of turns")
```

<b> Overall top scorer 

```python
pp = df.groupby("top_scorer")['highscore'].sum().sort_values(ascending=False)


fig = plt.figure(figsize = (10,5))
plt.xticks(rotation=20)
sns.barplot(x = pp.index , y = pp.values , palette='rainbow')
```

<b> 10 Best Bowling Figures

```python
df.sample(3)
```

```python
df['highest_wickets'] = df['best_bowling_figure'].apply(lambda x: x.split("--")[0])


df['highest_wickets'] = df['highest_wickets'].astype(int)


grp = df.groupby("best_bowling")['highest_wickets'].sum().sort_values(ascending = False).head(7)


fig = plt.figure(figsize = (10,5))
plt.xticks(rotation = 20)

sns.barplot(grp)
```
